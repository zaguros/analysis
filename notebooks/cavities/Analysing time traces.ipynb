```python
import numpy as np
import os.path
import pandas
import matplotlib
import scipy.signal

%matplotlib inline
import matplotlib.pyplot as plt
 
data_dir = "D:\measuring/data/20160523/"
```

## Interferometric trace

```python
file_name = "LONGSCAN000.csv"

filename = os.path.join(data_dir,file_name)

def get_data(filename):

    file = open(filename)
    for x in range(9):
            line = file.readline()
            if "HResolution" in line:
                TimeRes = float(line[-17:-4])
    file.close()

    subdiv = 1
    data = pandas.read_csv(os.path.join(data_dir,filename),header=None,skiprows = 16)
    print data
    data = data.drop(2,1)
    print data
    #data = data.drop(1,1)
    data = data.drop(0,1)
    #data = data.rolling_mean(window = 100)#.mean()
    #data = pandas.rolling_mean(data, window=10)
    data = data.as_matrix()[::subdiv]
    resn = TimeRes*subdiv
    t = resn*np.arange(len(data))    

    #NormMax = np.amax(data); NormMin = np.amin(data)
    #data = (np.array(data)-(NormMax+NormMin)/2)*(2./(NormMax-NormMin)) 
    return t,data,resn

t, data,resn = get_data(filename)
```

```python
freq_pow, power = scipy.signal.welch(data, axis = 0, fs = 1/resn,nperseg = 100000)

fig = plt.figure(figsize=(12, 3))

ax = plt.subplot(1,3,1)
plt.plot(freq_pow,power)
plt.xlim([0,10000])
#plt.ylim([0,0.004])
plt.xlabel('freq (Hz)')
plt.ylabel('Noise power (V**2/Hz)')

ax = plt.subplot(1,3,2)
        
plt.plot(freq_pow,power)
plt.xlim([1500,2500])
#plt.ylim([0,0.004])
plt.xlabel('freq (Hz)')

ax = plt.subplot(1,3,3) 
plt.plot(freq_pow,power)
plt.xlim([5000,6000])
#plt.ylim([0,0.8])
plt.xlabel('freq (Hz)')


plt.show()
plt.close(fig)
```

```python
min_freq = 10
max_freq = 10000

min_ind = np.argmin(np.abs(freq_pow-min_freq))
max_ind = np.argmin(np.abs(freq_pow-max_freq))

pow_int = scipy.integrate.trapz(power[min_ind:max_ind],freq_pow[min_ind:max_ind],axis=0)
full_pow = scipy.integrate.trapz(power,freq_pow,axis=0)

dc_freq = 10
dc_ind = np.argmin(np.abs(freq_pow-dc_freq))
print dc_ind

pow_dc = scipy.integrate.trapz(power[0:dc_ind],freq_pow[0:dc_ind],axis=0)

rel_dc_pow = pow_dc/full_pow
print 'Relative DC power: %.3f' %rel_dc_pow

rel_pow = pow_int/full_pow
print 'Relative in-band power: %.3f' %rel_pow

rel_band_pow = pow_int/(full_pow-pow_dc)
print 'Relative in-band power (minus DC): %.3f' %rel_band_pow

print 'Full noise power: %.3e V/sqrt(Hz)' %np.sqrt(full_pow)
```

```python
def get_noise_power(freq_pow,power):    
    min_freq = 10
    max_freq = 10000

    min_ind = np.argmin(np.abs(freq_pow-min_freq))
    max_ind = np.argmin(np.abs(freq_pow-max_freq))

    pow_int = scipy.integrate.trapz(power[min_ind:max_ind],freq_pow[min_ind:max_ind],axis=0)
    full_pow = scipy.integrate.trapz(power,freq_pow,axis=0)

    dc_freq = 10
    dc_ind = np.argmin(np.abs(freq_pow-dc_freq))

    pow_dc = scipy.integrate.trapz(power[0:dc_ind],freq_pow[0:dc_ind],axis=0)

    rel_dc_pow = pow_dc/full_pow

    rel_pow = pow_int/full_pow

    rel_band_pow = pow_int/(full_pow-pow_dc)

    return np.sqrt(full_pow)
 
```
