```python
import numpy as np
from analysis.lib.tools import toolbox as tb
from matplotlib import pyplot as plt
from analysis.lib.m2 import m2
from analysis.lib.tools import toolbox
from analysis.lib.fitting import fit,esr,common
from analysis.lib.tools import plot

def import_data(msmt_type,contains = None, N = 0, plot_raw_data = False, timestamp = None):
    """
    Imports QuTau data & integrates over ROI. 
    Outputs hdf5 file (f), sweeping parameter values (x) and counts (y). 
    NOTE: if msmt_type = DD, then x = total free evolution time & 'taus' returns the sweeped waiting time.
    Required input:
    - 'msmt_type': string specifying type of measurement (i.e. DESR, Rabi, DD)
    Optional input:
    - 'N': number of pi pulses in DD experiment (NOTE: msmt_type = 'DD' to use this input)
    - 'file_name': explicit file name including path of the data that is to be imported
    - 'plot_raw_data': Boolean that sets plotting of counts vs time for 1 data point on/off
    - 'timestamp': string specifying timestamp of data to be imported (NOTE: should supersede file_name but not yet implemented)
    """
    offset1 = 0.5
    offset2 = 225000
    Amplitude = 1500
    Phase = -10
    T1 = 0.5
    
    if timestamp != None:
        folder = toolbox.data_from_time(timestamp)
    elif contains != None:
        folder = toolbox.latest_data(contains=contains)
    print folder
        
    file_name = tb.measurement_filename(folder)
    f = h5py.File(file_name,'r')
    data_hist1 = f['QuTau_hist1'].value
    name=f.keys()[0]
    pts=len(f[name].attrs['sweep_pts'])
    binsize=f[name].attrs['summed_binsize']
    start=f[name].attrs['RO_start']
    stop=f[name].attrs['RO_stop']
    start_bin=int((start/.081/binsize))###???
    stop_bin=int((stop/.081/binsize))###???
    
    # Plot raw counts vs time
    if plot_raw_data:
        res_2D = []
        fig = plt.figure()
        ax = plt.subplot()
        for j in range(np.shape(data_hist1)[1]):
            dp_nr=j
            
            data=data_hist1[:,dp_nr]

            result = data[:(data.size // binsize) * binsize].reshape(-1, binsize).sum(axis=1)
            
            t=np.arange(len(result))*.081*binsize
            t_plot=t[start_bin:stop_bin]
            result_plot=result[start_bin:stop_bin]
            res_2D.append(result_plot)
            plt.plot(t_plot,result_plot)
            plt.xlabel('Time after sync [ns]')
            plt.ylabel('Counts')
#             plt.ylim([2000,4000])
            
            
        plt.show()
        plt.close('all')
        
        
        fig = plt.figure()
        ax = plt.subplot()
        plt.xlabel('Time after sync [ns]')
        plt.ylabel('Sweep point')
        plt.imshow(res_2D, cmap='hot', interpolation='nearest')
        
        plt.show()
        plt.close('all')
    S=[]
    y = np.zeros(pts)
    
    # Integrate over ROI
    for i in np.arange(pts):
        start_laser=np.where(data_hist1[:,i]==2)[0]
        y[i] += sum(data_hist1[start:int((stop/.081)),i]) # QuTau has binsize of 81 ps
        S.append(sum(data_hist1[start:int((stop/.081)),i]))
    if msmt_type == 'DD':
        if N == 0:
            raise Exception("Your DD measurement is specified with N = 0. This doesn't make sense.")
        taus = np.round(f[name].attrs['sweep_pts'],5)
        x = taus * 2 * N # free evolution time
    else:
        x = np.round(f[name].attrs['sweep_pts'],5)

    fig = plt.figure()
    plt.errorbar(x,y,yerr=sqrt(y),fmt='o',color='Crimson')
    
        
    # Plot formatting
    plt.xlabel(f[name].attrs['sweep_name'])
    plt.ylabel('PL')
    plt.title(folder)
    plt.savefig(os.path.join(folder,name+'.png'),format='png',bbox_inches='tight')
    plt.show()
    plt.close('all')
    
    if contains == 'Rabi':
        p0, fitfunc, fitfunc_str = common.fit_decaying_cos(offset1,offset2,Amplitude,Phase,T1)
        fit_result = fit.fit1d(x,y, None, p0=p0, fitfunc=fitfunc, do_print=False, ret=True)
        plot.plot_fit1d(fit_result, np.linspace(0,x[-1],201), ax, plot_data=True)
        plt.xlabel(f[name].attrs['sweep_name'])
        plt.ylabel('PL')
        plt.title(folder)
        plt.savefig(os.path.join(folder,name+'1.png'),format='png',bbox_inches='tight')
    
    if contains == tstamp:
        guess_offset=91000
        guess_amplitude = -5000
        guess_width = 0.0012
        guess_ctr =2.0001
        p0, fitfunc, fitfunc_str = common.fit_gauss(guess_offset,guess_amplitude,guess_width,guess_ctr)
        fit_result = fit.fit1d(x,y, None, p0=p0, fitfunc=fitfunc, do_print=False, ret=True)   
        plot.plot_fit1d(fit_result, np.linspace(2.698,x[-1],201), ax, plot_data=True)
        plt.xlabel(f[name].attrs['sweep_name'])
        plt.ylabel('PL')
        plt.title(folder)
        plt.savefig(os.path.join(folder,name+'1.png'),format='png',bbox_inches='tight')

    
#     x1=[]
#     y1=[]
#     for i in range(29):
#         x1.append(x[i])
#         y1.append(y[i])
#     x1=np.round(x1,5)    
#     fig = plt.figure()
#     plt.errorbar(x1,y1,yerr=sqrt(y1),fmt='o',color='Crimson')
    
#     o1 = 0.2
#     o2 = 225000
#     A = 4700
#     P = 0.5
#     T = 0.5
    
#     p0, fitfunc, fitfunc_str = common.fit_decaying_cos(o1,o2,A,P,T)
#     fit_result = fit.fit1d(x1,y1, None, p0=p0, fitfunc=fitfunc, do_print=False, ret=True)
#     plot.plot_fit1d(fit_result, np.linspace(0,x1[-1],201), ax, plot_data=True)
```
